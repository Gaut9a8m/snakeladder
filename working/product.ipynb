```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv(
'product_404.log',
usecols=[0, 1,3,5, 6, 7, 8, 10, 11, 17],
sep=r'\s(?=(?:[^"]*"[^"]*")*[^"]*$)(?![^\[]*\])',
delimiter=None, engine='python',
names=['Domain','IP','Time-Stamp','Response-Time','Host','Request','Status','Referer', 'User-Agent', 'X-IP']
)
```

```python
df
```

```python
tmp_df = df[['Request','User-Agent','X-IP', 'Referer']]
bot_df = tmp_df.loc[tmp_df['User-Agent'].str.contains('bot')==True]
google_bot_df = tmp_df.loc[tmp_df['User-Agent'].str.contains('GoogleBot', case=False)==True]
bing_bot_df = tmp_df.loc[tmp_df['User-Agent'].str.contains('BingBot',case=False)==True]
human_df = tmp_df.loc[tmp_df['User-Agent'].str.contains('bot')==False]
```

```python
bing_bot_df_c = bing_bot_df.loc[bing_bot_df['Request'].str.contains('-c[0-9]+', case=False)==True]
bing_bot_df_wc = bing_bot_df.loc[bing_bot_df['Request'].str.contains('-c[0-9]+', case=False)==False]
google_bot_df_c = google_bot_df.loc[google_bot_df['Request'].str.contains('-c[0-9]+', case=False)==True]
google_bot_df_wc = google_bot_df.loc[google_bot_df['Request'].str.contains('-c[0-9]+', case=False)==False]
human_df_c = human_df.loc[human_df['Request'].str.contains('-c[0-9]+', case=False)==True]
human_df_wc = human_df.loc[human_df['Request'].str.contains('-c[0-9]+', case=False)==False]
```
